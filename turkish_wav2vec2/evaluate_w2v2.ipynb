```python
#set the sampling rate
import librosa
audio, rate = librosa.load('/content/turkish.wav', sr=16000)
```

```python
audio, rate
```

```python
model = AutoModelForCTC.from_pretrained("cahya/wav2vec2-base-turkish-artificial")
processor = Wav2Vec2Processor.from_pretrained("cahya/wav2vec2-base-turkish-artificial")
tokenizer = Wav2Vec2Tokenizer.from_pretrained('cahya/wav2vec2-base-turkish-artificial')
```

```python
model = AutoModelForCTC.from_pretrained("ix502iv/wa2vec2-large-xls-r-colab_turkish")
processor = Wav2Vec2Processor.from_pretrained("ix502iv/wa2vec2-large-xls-r-colab_turkish")
tokenizer = Wav2Vec2Tokenizer.from_pretrained('ix502iv/wa2vec2-large-xls-r-colab_turkish')
```

```python
input_values = tokenizer(audio, return_tensors='pt').input_values
input_values
```

```python
# get and store logits
logits = model(input_values).logits
logits
```

```python
import torch
prediction = torch.argmax(logits, dim=-1)
prediction
```

```python
# erken parlamento seçimlerine gidilmesini bekliyor musunuz

transcription = tokenizer.batch_decode(prediction)[0]
transcription
```
